# JSON 로드 테스트

`data` 폴더에 있는 JSON 파일들이 정상적으로 로드되는지 테스트합니다.

```python
import json
import os
from pathlib import Path
```

```python
# 데이터 폴더 경로 설정
data_dir = Path('../data')
json_files = list(data_dir.glob('*.json'))

print(f"Found {len(json_files)} JSON files in {data_dir.absolute()}")
for f in json_files:
    print(f"- {f.name}")
```

```python
# 개별 파일 로드 테스트
def test_load_json(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        print(f"✅ Successfully loaded: {file_path.name} (Keys: {list(data.keys()) if isinstance(data, dict) else 'List'})")
        return True
    except Exception as e:
        print(f"❌ Failed to load: {file_path.name} - Error: {e}")
        return False

print("Testing all JSON files:")
results = [test_load_json(f) for f in json_files]

if all(results):
    print("\n✨ All JSON files are valid and loaded successfully!")
else:
    print(f"\n⚠️ {results.count(False)} file(s) failed to load.")
```

```python
# 특정 파일 상세 확인 (예: assets.json)
target_file = data_dir / 'assets.json'
if target_file.exists():
    with open(target_file, 'r', encoding='utf-8') as f:
        assets_data = json.load(f)
    print(f"\nContent of {target_file.name}:")
    print(json.dumps(assets_data, indent=2, ensure_ascii=False))
else:
    print(f"\n{target_file.name} not found.")
```
